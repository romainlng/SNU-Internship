```python
import torch
import numpy as np
```

TENSOR INITIALIZATION

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(np_array,"\n", x_np)
```

```python
x_ones = torch.ones_like(x_data)
print(x_ones)
x_rand = torch.rand_like(x_data, dtype=torch.float) #la ligne dtype est obligatoire, et le type doit etre un float
print(x_rand)
```

```python
shape=(1,2)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zero_tensor = torch.zeros(shape)
print(rand_tensor,ones_tensor,zero_tensor)
```

TENSORS ATTRIBUTES

```python
tensor = torch.rand(2,3)
print(tensor.shape)
print(tensor.dtype)
print(tensor.device)
```

TENSORS OPERATIONS

```python
# We move our tensor to the GPU if available
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device)
print(f"Device tensor is stored on: {tensor.device}")
```

```python
tensor = torch.ones(4,4)
tensor[:,2] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor,tensor,tensor], dim=1)
print(t1)
```

```python
tensor = torch.ones(3,2)
tensor[:,1] = 0
tensor *= 2
print(tensor) 
print(tensor.mul(tensor)) #mets au carrés chaque coeff du tenseur
print(tensor*tensor) #meme opération, différente syntax
```

```python
tensor = torch.ones(4,4)
tensor[:,1] = 0
print(tensor.matmul(tensor.T)) #multiplie 2 tensor, tensor.T doit etre la transposée
print(tensor@tensor.T) #autre syntax
```

```python
tensor = torch.rand(2,3)
print(tensor)
tensor.add_(5) #ajoute 5 à chaque coeff
print(tensor)
```

BRIDGE WITH NUMPY

```python
t = torch.ones(5)
print(t)
n = t.numpy()
print(n)
```

```python
t.add_(1)
print(t,n) #prouve la dépendance de t et n
```

```python
#Meme chose dans l'autre sens
n = np.ones(5)
print(n)
t = torch.from_numpy(n)
print(t)
```

```python
np.add(n,1,out=n)
print(t,n)
```
